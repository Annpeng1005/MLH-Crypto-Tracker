# Sprint 1: Live Market Data Ingestion

We are building a **Live Crypto Market Arbitrage Tracker**. Before we can build algorithms to analyze price drops and trigger alerts, we need a clean, structured dataset of the current market.

Your goal today is to pull live production data, stitch it together, and export a clean report.

---

## The Problem (Data Silos)
In production environments, APIs rarely give you everything in one payload to save bandwidth. The market data we need is split across two completely different API endpoints.

1. **The Roster (Metadata):** `https://api.coingecko.com/api/v3/coins/list`
   * *This contains the `id`, `symbol`, and `name` of every coin.*
2. **The Prices (Live Data):** `https://api.coingecko.com/api/v3/simple/price`
   * Use the `params` argument in the `get` method to filter for the following data points: `ids=bitcoin,ethereum,solana,ripple,dogecoin` and `vs_currencies=usd`.
   * *This contains the live USD price, mapped only to the coin `id`.*

---

## Your Acceptance Criteria (The Ticket)
Please write a Python script in the cell below that accomplishes the following:

1. Fetch the JSON data from both URLs.
2. Filter the massive roster down to just our five target coins: `bitcoin`, `ethereum`, `solana`, `ripple`, and `dogecoin`.
3. Write a nested loop (an Inner Join) to cross-reference the filtered roster with the live price data using the coin `id`.
4. Export a clean spreadsheet named `market_snapshot.csv` containing exactly three columns: `Coin Name`, `Symbol`, and `Current USD Price`.

```python
import requests
import csv

# Step 1: define the URLs
url_roster = "https://api.coingecko.com/api/v3/coins/list"
url_price = "https://api.coingecko.com/api/v3/simple/price"

#Step 2: define the target coin IDs
target_coins = ["bitcoin", "ethereum", "solana", "ripple", "dogecoin"]

#Step 3: make GET request to coin roster
roster_response = requests.get(url_roster)
roster_data = roster_response.json()
# print(roster_data)


```

```python


# Step 1: define the URLs
# url_roster = "https://api.coingecko.com/api/v3/coins/list"
url_price = "https://api.coingecko.com/api/v3/simple/price"

#Step 2: define the target coin IDs
target_coins = ["bitcoin", "ethereum", "solana", "ripple", "dogecoin"]

#Step 3: make GET request to coin roster
# roster_response = requests.get(url_roster)
# roster_data = roster_response.json()
# print(roster_data)
price_params= {'ids': 'bitcoin,ethereum,solana,ripple,dogecoin', 'vs_currencies': 'usd'}

price_response = requests.get(url_price, params=price_params)
price_data = price_response.json()

print(price_response.status_code)
print(price_data)
```

```python
#fixing data and dependency for the external data
#be clear with str. and int for sorting reason
```

```python
filter_coins=[]
for el in roster_data:
  # print(el)
  if el['id'] in target_coins:
    filter_coins.append(el)
print(filter_coins)

```

```python
#inner loop to cross reference the filter roster with the live pricing data
#match through their id information
filter_price=[]
for coin in filter_coins:
  # print(coin)
  for price_id in price_data:
    # print(price_id)
    if coin['id'] == price_id:
      # print(price_data[price_id]['usd'])
      # print(type(coin['name']))
      filter_price.append([coin['name'], coin['symbol'], price_data[price_id]['usd']]) #using list

print(filter_price)
```

```python

# Export a clean spreadsheet named market_snapshot.csv containing exactly three columns: Coin Name, Symbol, and Current USD Price.

headers = ['Coin Name', 'Symbol', 'Current USD Price']

with open('market_snapshot.csv', "w") as file:
  writer=csv.writer(file) #this will be for list and dictwriter is for dictionary
  writer.writerow(headers)
  writer.writerows(filter_price)
print("Report successfully generated")
```
